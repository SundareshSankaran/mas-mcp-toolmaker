# Microanalytics Services (MAS) - Model Context Protocol (MCP) Tool Maker

### This notebook helps you programmatically create a Model Context Protocol (MCP) server script to call published SAS Microanalytics Services modules discovered in a SAS Viya environment. The output is a Python script (in two flavours) which can be used to stand up an MCP server through stdio, Streamable HTTP and SAS Retrieval Agent Manager.

```python
# Hint: You can open this notebook in a virtual environment with a .env file containing your environment variables and load them through the dotenv package.

from dotenv import load_dotenv
load_dotenv()


import os
import requests
from requests.auth import HTTPBasicAuth
import getpass
import importlib
import sys

# The following modules are from the folder "modules" located at the same level as this notebook.

from modules.get_viya_info import *
from modules.get_module_info import *
from modules.script_stdio_server import *
from modules.script_ram_server import *
```

## Inputs

#### VIYA_ACCESS_TOKEN: An access token which helps you authenticate to SAS Viya

```python
viya_host = os.getenv("VIYA_HOST")
viya_access_token = os.getenv("VIYA_ACCESS_TOKEN")
```

## Generate a list of all modules
### Firstly, get the total number of modules at SAS MicroAnalytics Services (which may have come from published models or decisions)

```python
nbr_modules = get_initial_count(viya_host,viya_access_token)
```

```python
print(f"Number of Modules found: {nbr_modules}")
```

## Extract Module Properties

### Module properties are extracted after generating a list of all modules currently published to MAS

```python
modules_endpoints_properties = get_all_module_properties(viya_host = viya_host,access_token = viya_access_token, modules=get_all_modules(viya_host,viya_access_token,cnt_modules = nbr_modules))
```

#### An example of descriptions lifted from MAS modules' parent model / decision flow.

```python
modules_endpoints_properties["items"][0]["moduleProperties"]["description"]
```

## Consolidate all relevant metadata for these modules
### Get a list of endpoints, input and output contracts

```python
module_contracts = get_module_contracts(get_module_urls(modules_endpoints_properties, viya_host=viya_host))
```

#### An example of a module contract which represents the input to generate a MCP script from

```python
module_contracts[0]
```

## Generate MCP Servers (for stdio transport layer and RAM)
### Finally, enter a path where you wish to save a generated MCP server script file.  One is meant for use in Retrieval Agent Manager (RAM) and the other can be used with the stdio transport layer in applications such as Visual Studio Code.

```python
script_mcp_stdio_server("test_stdio_server.py",module_contracts = module_contracts)
```

```python
script_mcp_ram_server("test_ram_server.py",module_contracts = module_contracts)
```
