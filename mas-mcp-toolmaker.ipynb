## Imports

### Either import the entire package or import the main module, MASModule from the package

````python
import mas_mcp_toolmaker
# or
from mas_mcp_toolmaker import MASModule
````

### Initialise a MASModule object which builds from all available published modules in SAS Micro Analytic Service

````python
masm = MASModule()
````

### Or, you can initialise a MASModule object only for published decisions

````python
masm_decisions = MASModule(module_types=["Decision"])
````

### Or, you can initialise a MASModule object only for published models

````python
masm_models = MASModule(module_types=["Model"])
````

## Filter for your desired tools

### You can filter for relevant tools on the basis of endpoint or keywords found in description

````python
import copy
masm_filter_1 = copy.deepcopy(masm) 
masm_filter_1.filter_modules({'module_description': 'hmeq'})
masm_filter_1.list_modules()
````

### An example where you provide a list of keywords (default condition logic is OR at present, AND conditions will be added soon)

````python
masm_filter_2 = copy.deepcopy(masm) 
masm_filter_2.filter_modules({'module_description': ['hmeq','risk']})
masm_filter_2.list_modules()
````

## Create your MCP Server Script

### Create a script to run in stdio mode (for Visual Studio Code, Claude Code etc. and local runs)

````python
masm.script_stdio_server("test-mcp-stdio-server.py")
````

### Create a script to run in SAS Retrieval Agent Manager

````python
masm.script_ram_server()
````
